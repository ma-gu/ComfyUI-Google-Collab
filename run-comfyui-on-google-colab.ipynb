```python
# Clone ComfyUI custom nodes
# Install ComfyUI custom nodes
import pathlib, subprocess

root = pathlib.Path("/content/ComfyUI")
custom = root / "custom_nodes"
custom.mkdir(parents=True, exist_ok=True)

%cd /content/ComfyUI/custom_nodes

repos = [
    "https://github.com/Alectriciti/comfyui-adaptiveprompts.git",
    "https://github.com/scottmudge/ComfyUI-NAG.git",
    "https://github.com/melMass/comfy_mtb.git",
    "https://github.com/rgthree/rgthree-comfy.git",
    "https://github.com/kijai/ComfyUI-WanVideoWrapper.git",
    "https://github.com/Kosinkadink/ComfyUI-VideoHelperSuite.git",
    "https://github.com/Fannovel16/ComfyUI-Frame-Interpolation.git",
    "https://github.com/kijai/ComfyUI-KJNodes.git",
    "https://github.com/yolain/ComfyUI-Easy-Use.git",
    "https://github.com/Suzie1/ComfyUI_Comfyroll_CustomNodes.git",
    "https://github.com/city96/ComfyUI-GGUF.git",
    "https://github.com/Smirnov75/ComfyUI-mxToolkit.git",
]

# Клонирование
for url in repos:
    name = url.rsplit("/", 1)[-1].replace(".git", "")
    if (custom / name).exists():
        print(f"Skip, already exists: {name}")
        continue
    subprocess.run(["git", "clone", url], check=True)

# Установка requirements (если есть)
for url in repos:
    name = url.rsplit("/", 1)[-1].replace(".git", "")
    req = custom / name / "requirements.txt"
    if req.exists():
        print(f"Installing requirements for {name}")
        subprocess.run(["pip", "install", "-r", str(req)], check=True)
    else:
        print(f"No requirements.txt for {name}, skip")
```

```python
# Download Wan / GGUF / LoRA models for ComfyUI (Civitai через ?token=)

import os
import pathlib
import subprocess
from urllib.parse import urlparse, parse_qs, urlencode, urlunparse
from google.colab import userdata

# Берём токен из окружения (Colab Secrets / Variables)
#CIVITAI_TOKEN = os.environ.get("CIVITAI_TOKEN", "").strip()
CIVITAI_TOKEN = userdata.get("CIVITAI_TOKEN").strip()
if not CIVITAI_TOKEN:
    print("WARNING: CIVITAI_TOKEN не найден в os.environ – ссылки Civitai, требующие логин, могут не скачаться.")

root = pathlib.Path("/content/ComfyUI")

paths = {
    "vae": root / "models" / "vae",
    "text_encoders": root / "models" / "text_encoders",
    "diffusion": root / "models" / "diffusion_models",
    "loras": root / "models" / "loras",
}

for p in paths.values():
    p.mkdir(parents=True, exist_ok=True)

downloads = [
    # Wan VAE
    (
        "https://huggingface.co/OreX/Models/resolve/main/WAN/wan_2.1_vae.safetensors?download=true",
        paths["vae"] / "wan_2.1_vae.safetensors",
    ),
    # umt5 text encoder
    (
        "https://huggingface.co/OreX/Models/resolve/main/WAN/umt5_xxl_fp8_e4m3fn_scaled.safetensors?download=true",
        paths["text_encoders"] / "umt5_xxl_fp8_e4m3fn_scaled.safetensors",
    ),
    # Smooth Mix Wan 2.2 14B I2V 2.0 High (Civitai)
    (
        "https://civitai.red/api/download/models/2513182?fileId=2460270",
        paths["diffusion"] / "SmoothMix_Wan2.2_14B_I2V_2.0_High.safetensors",
    ),
    # Smooth Mix Wan 2.2 14B I2V 2.0 Low (Civitai)
    (
        "https://civitai.red/api/download/models/2513186?fileId=2460676",
        paths["diffusion"] / "SmoothMix_Wan2.2_14B_I2V_2.0_Low.safetensors",
    ),
    # GGUF High
    (
        "https://huggingface.co/QuantStack/Wan2.2-I2V-A14B-GGUF/resolve/main/HighNoise/Wan2.2-I2V-A14B-HighNoise-Q2_K.gguf?download=true",
        paths["diffusion"] / "Wan2.2-I2V-A14B-HighNoise-Q2_K.gguf",
    ),
    # GGUF Low
    (
        "https://huggingface.co/QuantStack/Wan2.2-I2V-A14B-GGUF/resolve/main/LowNoise/Wan2.2-I2V-A14B-LowNoise-Q2_K.gguf?download=true",
        paths["diffusion"] / "Wan2.2-I2V-A14B-LowNoise-Q2_K.gguf",
    ),
    # LoRA lightx2v
    (
        "https://huggingface.co/Kijai/WanVideo_comfy/resolve/709844db75d2e15582cf204e9a0b5e12b23a35dd/Lightx2v/lightx2v_I2V_14B_480p_cfg_step_distill_rank128_bf16.safetensors?download=true",
        paths["loras"] / "lightx2v_I2V_14B_480p_cfg_step_distill_rank128_bf16.safetensors",
    ),
]

def add_token_to_civitai_url(url: str, token: str) -> str:
    if not token:
        return url
    if "civitai.com" not in url and "civitai.red" not in url:
        return url

    parsed = urlparse(url)
    qs = parse_qs(parsed.query)
    # не перезаписываем, если токен уже вставлен руками
    if "token" in qs:
        return url

    qs["token"] = [token]
    new_query = urlencode(qs, doseq=True)
    new_parsed = parsed._replace(query=new_query)
    return urlunparse(new_parsed)

def build_aria2_cmd(url: str, out_path: pathlib.Path):
    # Для civitai добавляем token в сам URL
    url_with_token = add_token_to_civitai_url(url, CIVITAI_TOKEN)

    cmd = [
        "aria2c",
        "-x", "16",
        "-s", "16",
        "-o", out_path.name,
        "-d", str(out_path.parent),
        url_with_token,
    ]
    return cmd

for url, out_path in downloads:
    if out_path.exists():
        print(f"Already exists, skip: {out_path}")
        continue

    out_path.parent.mkdir(parents=True, exist_ok=True)
    print(f"\nDownloading {url} -> {out_path}")

    cmd = build_aria2_cmd(url, out_path)
    print("Command:", " ".join(cmd))
    try:
        # Важно: для ссылок с '&' aria2c уже видит всё как один аргумент,
        # т.к. мы не даём shell и передаём список аргументов subprocess-у.
        subprocess.run(cmd, check=True)
    except subprocess.CalledProcessError as e:
        print(f"Download failed for {url}: {e}")
```
